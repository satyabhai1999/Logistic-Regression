Question 2. What are the key assumptions of Simple Linear Regression?

Ans-

1. Linearity

The relationship between the independent variable (X) and the dependent variable (Y) must be linear.

This means a straight line can best describe how Y changes with X.

Example: As study hours increase, marks increase in a roughly straight-line pattern.

2. Independence of Errors

The residuals (errors) â€” the differences between observed and predicted values â€” should be independent of each other.

This means the error for one observation doesnâ€™t depend on the error for another.

- Violation Example: Time series data where errors are correlated over time (autocorrelation).

3. Homoscedasticity (Constant Variance of Errors)

The variance of errors should be constant across all levels of X.

The spread of residuals should remain roughly the same, regardless of the value of X.

- Violation Example: When errors increase as X increases (called heteroscedasticity).

4. Normality of Errors

The residuals should be normally distributed (especially important for hypothesis testing and confidence intervals).

This can be checked with a histogram or Q-Q plot of residuals.

5. No or Minimal Multicollinearity

In Simple Linear Regression, there is only one independent variable, so this usually doesnâ€™t apply.

But in general regression, it means predictors should not be highly correlated with each other.

6. No Measurement Error in X

The independent variable (X) is assumed to be measured accurately without error.

Errors in measuring X can bias the slope estimate (b).


| **Assumption**       | **Meaning**                     | **Why It Matters**                     |
| -------------------- | ------------------------------- | -------------------------------------- |
| Linearity            | Y changes linearly with X       | Ensures model form is correct          |
| Independence         | Errors are independent          | Prevents misleading significance tests |
| Homoscedasticity     | Constant error variance         | Ensures reliability of predictions     |
| Normality            | Errors are normally distributed | Validates hypothesis tests             |
| No measurement error | X measured accurately           | Keeps slope estimate unbiased          |

Question 8: What is the purpose of the R-squared metric in regression analysis?

Ans-

R-squared (RÂ²) in Regression Analysis

Definition:
R-squared is a metric that shows how well the independent variable(s) explain the variation in the dependent variable.

In simple words, it tells us how much of the change in Y is explained by X.

Key Points:

Range: 0 to 1

0: X does not explain any variation in Y

1: X explains all the variation in Y perfectly

Interpretation:

RÂ² = 0.8 â†’ 80% of the variation in Y is explained by X

The remaining 20% is due to other factors or randomness

Purpose of R-squared:

Measure Model Fit:
Shows how well the regression line fits the data.

Compare Models:
Higher RÂ² generally indicates a better model for the same dataset.

Understand Variability:
Helps to see how much of Yâ€™s variability can be predicted from X.

In short:

R-squared tells us how good the regression model is at explaining the changes in the dependent variable. The closer it is to 1, the better the model fits the data.

```python
""" Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)"""

# solution-

# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# X = independent variable (e.g., hours studied)
# Y = dependent variable (e.g., marks)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape to 2D array
Y = np.array([50, 55, 65, 70, 75])

# Create the Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(X, Y)

# Get slope (b) and intercept (a)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (b): {slope}")
print(f"Intercept (a): {intercept}")
```
